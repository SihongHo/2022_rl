```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple
```

```python
n_groups = 1

means_rmaac1 = (100-0.643*24)
std_rmaac1 = (0.383)

means_rmaac2 = (100-0.61*24)
std_rmaac2 = (0.376)

means_m3ddpg = (100-0.43*24)
std_m3ddpg = (0.295)

means_maddpg = (100-0.462*24)
std_maddpg = (0.328)

fig, ax = plt.subplots(figsize=(3, 6))

index = np.arange(n_groups)
bar_width = 0.15

opacity = 0.5
error_config = {'ecolor': '0.1'}

rects1 = ax.bar(index, means_rmaac1, bar_width,
                alpha=opacity, color='r',
                yerr=std_rmaac1, error_kw=error_config, capsize = 3,
                label='RM1')

rects1_2 = ax.bar(index + bar_width * 1, means_rmaac2, bar_width,
                alpha=opacity, color='orange',
                yerr=std_rmaac2, error_kw=error_config, capsize = 3,
                label='RM2')

rects2 = ax.bar(index + bar_width * 2, means_m3ddpg, bar_width,
                alpha=opacity, color='g',
                yerr=std_m3ddpg, error_kw=error_config, capsize = 3,
                label='M3')

rects2 = ax.bar(index + bar_width * 3, means_maddpg, bar_width,
                alpha=opacity, color='b',
                yerr=std_maddpg, error_kw=error_config, capsize = 3,
                label='MA')

# ax.set_xlabel('Scenario')
ax.set_ylabel('Mean Episode Testing Rewards', fontsize=20)
ax.set_title('No Noise', fontsize=20)
ax.set_xticks(index + 1.5 * bar_width)
ax.set_xticklabels(('simple reference',), fontsize=20)
ax.tick_params(axis='y', labelsize=16)
ax.legend(fontsize=16)
# ax.set_xlim([xmin, xmax])
ax.set_ylim([70, 99])

fig.tight_layout()
plt.show()
```
