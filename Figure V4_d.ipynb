```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple

n_groups = 3

means_rmaac = (100-24.552, 100-65.28, 100-7.008)
std_rmaac = (0.872, 0.322, 2.304)

means_m3ddpg = (100-26.904, 100-66.264, 100-8.328)
std_m3ddpg = (1.012, 0.349, 2.711)

means_maddpg = (100-25.992, 100-67.008, 100-8.016)
std_maddpg = (0.976, 0.359, 2.811)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.25

opacity = 0.5
error_config = {'ecolor': '0.1'}

rects1 = ax.bar(index, means_rmaac, bar_width,
                alpha=opacity, color='r',
                yerr=std_rmaac, error_kw=error_config, 
                label='RMAAC')

rects2 = ax.bar(index + bar_width, means_m3ddpg, bar_width,
                alpha=opacity, color='g',
                yerr=std_m3ddpg, error_kw=error_config,
                label='M3DDPG')

rects2 = ax.bar(index + bar_width * 2, means_maddpg, bar_width,
                alpha=opacity, color='b',
                yerr=std_maddpg, error_kw=error_config,
                label='MADDPG')

ax.set_xlabel('Scenario')
ax.set_ylabel('Reward')
ax.set_title('Gaussion noise format')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(('simple_reference', 'simple_spread', 'simple_tag'))
ax.legend(loc = "upper center")

fig.tight_layout()
plt.show()
```
